Quick SLIC script from https://www.pyimagesearch.com/2014/07/28/a-slic-superpixel-tutorial-using-python/

```python
from skimage.segmentation import slic, mark_boundaries
from skimage.util import img_as_float
from skimage import io
import matplotlib.pyplot as plt
```

```python
# img_pth = '~/Desktop/middleburry_data_assembled/Phil_outputs/t10r10/Adirondack-perfect_im0.png'
img_pth = '~/Desktop/middleburry_data_assembled/Phil_outputs/tri_test.png'

img = io.imread(img_pth)
img = img / np.max(img)
img = img_as_float(img)

type(img), img.shape, np.min(img), np.max(img)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20,20))

ax=ax.flatten()
for i, numSegments in enumerate([20, 100,200,300]):
    segments = slic(img, n_segments=numSegments, sigma=5)

    ax[i].imshow(mark_boundaries(img, segments))
    ax[i].set_xlabel(f'Superpixels -- {numSegments} segments')

plt.tight_layout()
# plt.savefig('tri_out.png')
plt.show()
```
